```python
import yfinance as yf

# Define the ticker symbol
ticker_symbol = "TSLA"

# Create a Ticker object
ticker = yf.Ticker(ticker_symbol)

# Fetch historical market data for the last 30 days
historical_data = ticker.history(period="6mo", interval="1d")  # data for the last month

# Display a summary of the fetched data
print(f"Summary of Historical Data for {ticker_symbol}:")
print(historical_data[['Open', 'High', 'Low', 'Close', 'Volume']])
```

```python
fig = go.Figure(
    data=[
        go.Candlestick(
            x=historical_data.index,          # dates
            open=historical_data["Open"],     # map columns
            high=historical_data["High"],
            low=historical_data["Low"],
            close=historical_data["Close"],
            name=ticker_symbol
        )
    ]
)

# 3. Add a title and keep it interactive (zoom/pan enabled)
fig.update_layout(
    title=f"{ticker_symbol} Candlestick Chart - Last 6 Months",
    xaxis_title="Date",
    yaxis_title="Price (USD)",
    xaxis_rangeslider_visible=True  # slider helps zoom to individual candles
)

fig.show()
```
