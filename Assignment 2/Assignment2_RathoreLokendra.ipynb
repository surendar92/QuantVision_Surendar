```python
import yfinance as yf
import plotly.graph_objects as go
import pandas as pd
import numpy as np
from datetime import datetime,timedelta

symbol = "TSLA"
ticker = yf.Ticker(symbol)
data = ticker.history(period = "6mo", interval = '1d')

candelstick = go.Candlestick(
    x = data.index,
    open = data['Open'],
    high = data['High'],
    low = data['Low'],
    close = data['Close'],
    name = 'Candlestick'
)

fig = go.Figure(data = [candelstick])
chart_title = f'{symbol} Stock Price - Last 6 Months(Daily Candlestick)'

fig.update_layout(
    title = chart_title,
    xaxis_title = 'Date',
    yaxis_title = 'Price(Usd)',
    xaxis_rangeslider_visible = False,
    hovermode = 'closest'
)
fig.show()
```
