```python
import numpy as np
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
import plotly.io as pio
```

```python
pio.renderers.default = "notebook_connected"
```

```python
df = yf.download(
    "AAPL",
    period="6mo",
    interval="1d",
    group_by="column",
    auto_adjust=False
)
df.columns = df.columns.get_level_values(0)
df = df[df["Volume"] > 0]
```

```python
print(df[["Open", "High", "Low", "Close"]].head())
print(df[["Open", "High", "Low", "Close"]].describe())
print(df.index.dayofweek.value_counts())
```

```python
x = df.index.astype(str) 
open_  = df["Open"].to_numpy()
high_  = df["High"].to_numpy()
low_   = df["Low"].to_numpy()
close_ = df["Close"].to_numpy()
ymin = df["Low"].min()
ymax = df["High"].max()
padding = 0.02 * (ymax - ymin)

```

```python
fig = go.Figure(
    go.Candlestick(
        x=df.index.astype(str),  
        open=df["Open"],
        high=df["High"],
        low=df["Low"],
        close=df["Close"]
    )
)
fig.update_layout(
    height=1000
)


fig.update_layout(
    title="Apple (AAPL) â€” Last 6 Months",
    xaxis_title="Date",
    yaxis_title="Price (USD)",
    xaxis_rangeslider_visible=True,
    xaxis=dict(type="category") 
)
fig.update_layout(
    yaxis=dict(
        range=[ymin - padding, ymax + padding]
    )
)

fig.show()

```
