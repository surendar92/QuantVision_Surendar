```python
import yfinance as yf
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
ticker_symbol = "TSLA"
ticker = yf.Ticker(ticker_symbol)
historical_data = ticker.history(period='6mo', interval ='1d')
print(f"Historical data for {ticker_symbol} for 6 months is :", historical_data)
```

```python
fig = go.Figure(data=go.Candlestick(x = historical_data.index,
                                    open = historical_data['Open'],
                                    close = historical_data['Close'],
                                    low = historical_data['Low'],
                                    high = historical_data['High'],
                                    increasing = dict(line=dict(color = 'Green')),
                                    decreasing = dict(line = dict(color = 'Red'))))

fig.update_layout(
    title=f'{ticker_symbol} Candlestick Chart',
    xaxis_title='Date',
    yaxis_title='Price(USD) in millions',
    xaxis_rangeslider_visible=True)
fig.show()
```
