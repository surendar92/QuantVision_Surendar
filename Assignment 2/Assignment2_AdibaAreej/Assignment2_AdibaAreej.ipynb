```python
import yfinance as yf

ticker_symbol = "GS"

ticker = yf.Ticker(ticker_symbol)

historical_data = ticker.history(period="6mo", interval="1d")

print(f"Summary of Historical Data for {ticker_symbol}:")
print(historical_data[['Open', 'High', 'Low', 'Close', 'Volume']])


```

```python
fig = go.Figure(data=go.Candlestick(
            x=historical_data.index,
            open=historical_data["Open"],
            high=historical_data["High"],
            low=historical_data["Low"],
            close=historical_data["Close"],
            name=ticker_symbol))


fig.update_layout(
    title=f"{ticker_symbol} Candlestick Chart - Last 6 Months History",
    xaxis_title="Date",
    yaxis_title="Stock Price (USD)",
    xaxis_rangeslider_visible=True,
    template = 'plotly_dark')

fig.show()
```
