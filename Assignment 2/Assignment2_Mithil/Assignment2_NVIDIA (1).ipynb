```python
import yfinance as yf
import plotly.graph_objects as go
ticker_symbol = 'NVDA'
ticker = yf.Ticker(ticker_symbol)
historical_data = ticker.history(period='6mo', interval='1d')
fig = go.Figure(data=[go.Candlestick(x=historical_data.index,  open=historical_data['Open'], high=historical_data['High'], low=historical_data['Low'],
    close=historical_data['Close'])])
fig.update_layout(title='Candlestick chart of NVIDIA', xaxis_title = 'Dates', yaxis_title = 'Price' )
fig.show()

```
