```python
import yfinance as yf
import plotly as px
```

```python
ticker_symbol = 'NVDA'
ticker =  yf.Ticker(ticker_symbol)
```

```python
from pandas._libs import interval
historical_data = ticker.history(period = '6mo',interval = '1d')
```

```python
historical_data
```

```python
import plotly.graph_objs as go
import plotly.subplots as sp
increasing_color = 'green'
decreasing_color = 'red'

get_volume_color = lambda close_price, open_price: increasing_color if (close_price - open_price) > 0 else decreasing_color
volume_colors = [
    get_volume_color(close_price, open_price)
    for close_price, open_price in zip(historical_data['Close'], historical_data['Open'])
]

fig = sp.make_subplots(rows = 2 ,cols = 1 , shared_xaxes= True , vertical_spacing= 0.02)

fig.add_trace(go.Candlestick(x = historical_data.index,
                                       open = historical_data['Open'],
                                       high = historical_data['High'],
                                       close = historical_data['Close'],
                                       low = historical_data['Low']
                                       ),row = 1, col = 1)
fig.add_trace(go.Bar(x = historical_data.index, y = historical_data['Volume'], marker_color = volume_colors),row = 2, col = 1)
fig.update_layout(xaxis_rangeslider_visible = False, template = 'plotly_dark')
fig.show()
```
