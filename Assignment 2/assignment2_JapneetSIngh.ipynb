```python
import yfinance as yf
```

```python
import plotly.graph_objects as go
```

```python
ticker_symbol = "META"
```

```python
ticker = yf.Ticker(ticker_symbol)
historical_data = ticker.history(period="6mo")
```

```python
print(historical_data)
```

```python
fig = go.Figure(data=[go.Candlestick(x=historical_data.index, open=historical_data['Open'], high=historical_data['High'], low=historical_data['Low'], close=historical_data['Close'])])
```

```python
fig.show()
```
