```python
import yfinance as yf
```

```python
import plotly.graph_objects as ply
```

```python
df=yf.download(tickers="TSLA",period="6mo", interval="1d")
df.head()
```

```python
df.columns=["Close","High","Low","Open","Volume"]
df.head()
```

```python
df=df.reset_index() #to make Date as a column in the dataframe
df.head()
```

```python
graph=ply.Figure(data=[ply.Candlestick(x=df['Date'],open=df['Open'],high=df['High'],low=df['Low'],close=df['Close'])])
graph.show()
```

```python
graph.update_layout(title="Tesla Stock Price Analysis",title_x=0.5)
```
