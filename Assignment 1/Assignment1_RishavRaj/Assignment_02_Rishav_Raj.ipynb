```python
import yfinance as yf
import plotly.graph_objects as go
import pandas as pd

ticker_symbol = "AAPL"  #Apple
ticker = yf.Ticker(ticker_symbol)

historical_data = ticker.history(period="6mo",interval="1d")


fig = go.Figure(data=[go.Candlestick(
    x=historical_data.index,
    open=historical_data['Open'],
    high=historical_data['High'],
    low=historical_data['Low'],
    close=historical_data['Close']
)])


fig.update_layout(
    title="AAPL Chart",
    xaxis_rangeslider_visible=False
)


fig.show()
```

1} A ***Bearish Engulfing pattern*** can be seen on date Sep 17 and Sep 18 where a bullish candelstick is engulfed by a bearish one.
It is a strange case where the price didn't decline but went upward.

2} A ***Doji Cross*** is also seen on Sep 17 showing that the opening and closing remained the same on that day. It follows its definition that the market will slowdown or uncertainity.
