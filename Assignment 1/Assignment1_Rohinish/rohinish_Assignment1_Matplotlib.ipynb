**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Main data
x = np.arange(0, 101)
y = x**2

plt.figure(figsize=(10, 6))

# Main plot
plt.plot(x, y, 'b')
plt.xlabel('X')
plt.ylabel('Z')

# Inset axes (zoom window)
inset = plt.axes([0.30, 0.45, 0.28, 0.32])   # (left, bottom, width, height)

# Correct zoom line: y goes from 40 to 45 as x goes 20 to 22
x_zoom = np.array([20, 22])
y_zoom = np.array([40, 45])

inset.plot(x_zoom, y_zoom, 'b')
inset.set_title("zoom")
inset.set_xlabel("X")
inset.set_ylabel("Y")

# Set inset axis limits to match your screenshot
inset.set_xlim(20.0, 22.0)
inset.set_ylim(30, 50)

plt.show()

```
