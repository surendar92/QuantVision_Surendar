```python
#describe the data through statistics
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("Iris.csv")

# List of species
species = df["Species"].unique()

# ------------------ Plot 1 ------------------
plt.figure(figsize=(6,4))
for s in species:
    temp = df[df["Species"] == s]
    plt.scatter(temp["SepalLengthCm"], temp["SepalWidthCm"], label=s)

plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("Scatter Plot: SepalLength vs SepalWidth (Colored by Species)")
plt.legend()
plt.grid(True)
plt.show()

# ------------------ Plot 2 ------------------
plt.figure(figsize=(6,4))
for s in species:
    temp = df[df["Species"] == s]
    plt.scatter(temp["PetalLengthCm"], temp["PetalWidthCm"], label=s)

plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")
plt.title("Scatter Plot: PetalLength vs PetalWidth (Colored by Species)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from IPython.display import display, HTML

df_sepal = df[['SepalLengthCm', 'SepalWidthCm']]
df_petal = df[['PetalLengthCm', 'PetalWidthCm']]

# Convert DataFrames to HTML
html_sepal = df_sepal.to_html()
html_petal = df_petal.to_html()

# Combine HTML strings with CSS to display them side-by-side
html_combined = f"<div style='display: flex;'><div>{html_sepal}</div><div style='margin-left: 20px;'>{html_petal}</div></div>"

# Display the combined HTML
display(HTML(html_combined))
# From taking multiple columns from the dataset
```
